```python
def team_score_diffs(
        team_input: str, gamemode_input: str, map_input = "All"
):
    filtered_df = \
        cdlDF[(cdlDF["gamemode"] == gamemode_input) & \
                (cdlDF["team"] == team_input)] \
                [["match_id", "map_name", "score_diff"]].drop_duplicates()
    
    return filtered_df
```
